```python
lending_new1 = pd.get_dummies(data=lending_new1,columns= ['home_ownership','verification_status','purpose','loan_status'])
```

```python
lending_new1 = lending_new1.dropna(axis = 0)
```

```python
y = lending_new1['int_rate']
x = lending_new1.drop('int_rate',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
y_pred1 = model1.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score, mean_absolute_error
Variance_score = explained_variance_score(y_test, y_pred)
Variance_score1 = explained_variance_score(y_train, y_pred1)
print ('Variance score_train : %.2f' %Variance_score1)
print('Variance score_test : %.2f' %Variance_score)


#Mean Absolute Error
MAE = mean_absolute_error(y_test, y_pred)
MAE1 = mean_absolute_error(y_train, y_pred1)
print ('Mean Absolute Error_train : %.2f' %MAE1)
print ('Mean Absolute Error_test : %.2f' %MAE)


#Root Mean Squared Error
RMSE = mean_squared_error(y_test,y_pred)**0.5
RMSE1 = mean_squared_error(y_train,y_pred1)**0.5
print ('Mean Squared Error_train : %.2f' %RMSE1)
print ('Mean Squared Error_test : %.2f' %RMSE)
#RÂ² score, the coefficient of determination
r2s = r2_score(y_test, y_pred)
r3s = r2_score(y_train,y_pred1)
print ('R2  score train: %.2f' %r3s)
print ('R2  score test: %.2f' %r2s)
```
